## 7. VProfile Project

Project: Set up a Web Application Stack in a local machine

### About the Project

Multi-tier Web Application Stack

Setup on Laptop/Desktop (Local)

Reasons for the project:

* Set a baseline for our upcoming projects.

* Help you set up any project locally.

The objective:

* Get a local setup that is automated, repeatable and based on Infrastructure as Code(IAC)

Tools:

* Hypervisor: Oracle VM VirtualBox

* Automation: Vagrant

* Command Line Tool (CLI): Git Bash

* IDE: Sublime, VS Code, etc.

### Architecture of the Project

We will be setting up a web application -> A social networking site <ins>written in Java</ins>.

"Stack" is a collection of services working together to create an experience: web or app

Project stack:



* Users connect to the web through an IP address (for now)

* The IP address belongs to a load balancer: *NGINX* (it's similar to apache or httpd)

* Then, NGINX routes the request to an *APACHE TOMCAT* service: a Java Web Application service. 

    - This is where the Java Application will be placed. (commonly used)
    
    - If the application needs additional storage, you can use NFS servers*.

    - Users will log in to the application in TOMCAT

    - The login request will be sent to the *MEMCACHED* service: A database caching. Saves information stored in MySQL for future requests.

    - Finally, the request is connected to a *MySQL* server, where the user information will be stored

* Connected to TOMCAT, there will be a message broker, aka Queuing Agent, which is used to connect two applications together: In this case, we will use RabbitMQ (it's going to be dummy for this project)


*The Network File System (NFS) is a mechanism for storing files on a network. It is a distributed file system that allows users to access files and directories located on remote computers and treat those files and directories as if they were local.

- Enables multiple users to access files

- Uses TCP/IP or the User Datagram Protocol

- Used to deploy mission-critical applications: Oracle DBs or VMs

### Flow of Execution

1. Set up tools mentioned in the prerequisites

2. Clone source code

3. CD into the vagrant directory

4. Bring up VMs

5. Validate that the VMs are up and running and able to interact with each other

6. Set up all the services:

    - MySQL

    - Memcached

    - Rabbit MQ

    - Tomcat

    - Nginx

    - App build and deploy

7. Verify from browser
